```python
# Visualization: Python, Matplotlib & seaborn
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# CREATE DUMMY DATA
# =========================
data = {
    "month": ["Jan", "Feb", "Mar", "Apr", "May", "Jun"],
    "sales": [120, 150, 170, 160, 190, 210],
    "profit": [30, 45, 55, 50, 65, 80],
    "region": ["West", "West", "East", "East", "West", "East"]
}

df = pd.DataFrame(data)

# =========================
# 1. LINE CHART (Sales Trend)
# =========================
plt.figure()
plt.plot(df["month"], df["sales"])
plt.xlabel("Month")
plt.ylabel("Sales")
plt.title("Monthly Sales Trend")
plt.show()

# =========================
# 2. BAR CHART (Profit per Month)
# =========================
plt.figure()
plt.bar(df["month"], df["profit"])
plt.xlabel("Month")
plt.ylabel("Profit")
plt.title("Monthly Profit Comparison")
plt.show()

# =========================
# 3. SEABORN BOXPLOT (Sales by Region)
# =========================
plt.figure()
sns.boxplot(data=df, x="region", y="sales")
plt.title("Sales Distribution by Region")
plt.show()

fig, ax = plt.subplots(figsize=(8, 5))
sns.boxplot(data=df, x="region", y="sales", ax=ax)
ax.set_title("Sales Distribution by Region")
plt.tight_layout()
plt.show()

```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# =========================
# CREATE DUMMY DATA
# =========================
data = {
    "week": ["W1", "W2", "W3", "W4", "W5", "W6"],
    "orders": [45, 60, 55, 70, 85, 90],
    "revenue": [4.5, 6.2, 5.8, 7.5, 9.1, 9.8],
    "channel": ["Online", "Offline", "Online", "Offline", "Online", "Offline"]
}

df = pd.DataFrame(data)

# =========================
# 1. LINE CHART (Order Trend)
# =========================
plt.figure(figsize=(8, 5))
plt.plot(df["week"], df["orders"])
plt.xlabel("Week")
plt.ylabel("Orders")
plt.title("Weekly Order Trend")
plt.tight_layout()
plt.show()

# =========================
# 2. BAR CHART (Revenue per Week)
# =========================
plt.figure(figsize=(8, 5))
plt.bar(df["week"], df["revenue"])
plt.xlabel("Week")
plt.ylabel("Revenue (Million)")
plt.title("Weekly Revenue Comparison")
plt.tight_layout()
plt.show()

# =========================
# 3. SEABORN BOXPLOT (Orders by Channel)
# =========================
plt.figure(figsize=(8, 5))
sns.boxplot(data=df, x="channel", y="orders")
plt.xlabel("Channel")
plt.ylabel("Orders")
plt.title("Order Distribution by Channel")
plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# =========================
# CREATE DUMMY DATA
# =========================
data = {
    "day": ["Mon", "Tue", "Wed", "Thu", "Fri", "Sat", "Sun"],
    "visitors": [320, 340, 360, 390, 420, 500, 480],
    "conversion_rate": [2.1, 2.3, 2.4, 2.6, 2.8, 3.1, 3.0],
    "device": ["Mobile", "Desktop", "Mobile", "Desktop", "Mobile", "Mobile", "Desktop"]
}

df = pd.DataFrame(data)

# =========================
# 1. LINE CHART (Visitor Trend)
# =========================
plt.figure(figsize=(8, 5))
plt.plot(df["day"], df["visitors"])
plt.xlabel("Day")
plt.ylabel("Visitors")
plt.title("Daily Visitor Trend")
plt.tight_layout()
plt.show()

# =========================
# 2. BAR CHART (Conversion Rate per Day)
# =========================
plt.figure(figsize=(8, 5))
plt.bar(df["day"], df["conversion_rate"])
plt.xlabel("Day")
plt.ylabel("Conversion Rate (%)")
plt.title("Daily Conversion Rate")
plt.tight_layout()
plt.show()

# =========================
# 3. SEABORN BOXPLOT (Visitors by Device)
# =========================
plt.figure(figsize=(8, 5))
sns.boxplot(data=df, x="device", y="visitors")
plt.xlabel("Device")
plt.ylabel("Visitors")
plt.title("Visitor Distribution by Device")
plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np

# =========================
# CREATE DUMMY DATA
# =========================
np.random.seed(42)

data = {
    "ad_spend": np.random.randint(10, 100, 20),
    "sales": np.random.randint(50, 300, 20),
    "customer_satisfaction": np.random.normal(7, 1, 20),
    "month": ["Jan", "Feb", "Mar", "Apr", "May"] * 4
}

df = pd.DataFrame(data)

# =========================
# 1. SCATTER PLOT (Ad Spend vs Sales)
# =========================
plt.figure(figsize=(8, 5))
plt.scatter(df["ad_spend"], df["sales"])
plt.xlabel("Ad Spend")
plt.ylabel("Sales")
plt.title("Relationship Between Ad Spend and Sales")
plt.tight_layout()
plt.show()

# =========================
# 2. HISTOGRAM (Customer Satisfaction)
# =========================
plt.figure(figsize=(8, 5))
plt.hist(df["customer_satisfaction"], bins=8)
plt.xlabel("Customer Satisfaction Score")
plt.ylabel("Frequency")
plt.title("Distribution of Customer Satisfaction")
plt.tight_layout()
plt.show()

# =========================
# 3. SEABORN HEATMAP (Monthly Average Sales)
# =========================
pivot_table = df.pivot_table(
    values="sales",
    index="month",
    aggfunc="mean"
)

plt.figure(figsize=(8, 5))
sns.heatmap(pivot_table, annot=True, fmt=".0f")
plt.title("Average Sales by Month")
plt.tight_layout()
plt.show()
```
